### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read file

```python
file = pd.read_csv('train.csv')
file.head()
```

```python
file.describe()
```

### Missing Values

```python
# Drop rows with missing target values
file = file.dropna(subset=['SalePrice'])

# Select numeric columns only for simplicity
numeric_file = file.select_dtypes(include=[np.number])

# Drop columns with too many missing values
numeric_file = numeric_file.dropna(axis=1)
```

### Features and Target

```python
x = numeric_file.drop("SalePrice", axis=1)
y = numeric_file["SalePrice"]
```

### Spliting dataset into Training and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Train Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=200, random_state=42)
regressor.fit(x_train, y_train)
```

### Predictions and Evaluations

```python
from sklearn.metrics import mean_absolute_error, r2_score
y_pred = regressor.predict(x_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"RÂ² score: {r2:.3f}")
print(f"Mean Absolute Error: {mae:.2f}")
```

### Plotting results

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual SalePrice")
plt.ylabel("Predicted SalePrice")
plt.title("Actual vs Predicted House Prices (Random Forest)")
plt.show()
```

### Feature Importance Plot

```python
importances = pd.Series(regressor.feature_importances_, index=x.columns)
top_features = importances.nlargest(10)

plt.figure(figsize=(8,6))
top_features.plot(kind='barh')
plt.title("Top 10 Important Features in Price Prediction")
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.show()
```
